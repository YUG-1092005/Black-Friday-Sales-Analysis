## Analyzing the product gender

```python
data = pd.DataFrame({"Ratio": [len(df[df["Gender"] == "M"]), len(df[df["Gender"] == "F"])]}, index=["Male","Female"])
data.plot.pie(y="Ratio", autopct="%.2f", title = 'Male Female Ratio(%)')
```

```python
df.groupby("Gender").size()
```

```python
df.groupby("Gender").size().plot(kind="pie", title="Male Female Ratio(%)", autopct="%.2f")
```

```python
df.groupby("Gender").size().plot(kind="bar", title="Male Female Ratio")
```

```python
df.groupby("Gender")["Purchase"].sum()
```

```python
df.groupby("Gender")["Purchase"].sum().plot(kind="pie", autopct="%.2f", title="Purchase of Male and Female(%)")
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,5))

purchase_data = df.groupby("Gender")["Purchase"].mean()

purchase_data.plot(kind="bar", ax=axes[0], title="Purchase per Male and Female", color=["pink", "blue"])
axes[0].set_ylabel("Amount of Purchases")

purchase_data.plot(kind="pie", ax=axes[1], title="Purchase per Male and Female(%)", autopct="%.2f")
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()
```
